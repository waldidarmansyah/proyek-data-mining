```
import pandas as pd

```

```
data = pd.read_csv(r'polusi.csv')
data
```

```
data.columns
```

```
data.rename(columns={'Country':'NEGARA', 'City':'KOTA', 'AQI Value':'NILAI_AQI', 'AQI Category':'KATEGORI_AQI', 'CO AQI Value':'NILAI_AQI_CO',
       'CO AQI Category':'KATEGORI_AQI_CO', 'Ozone AQI Value':'NILAI_AQI_OZON', 'Ozone AQI Category':'KATEGORI_AQI_OZON',
       'NO2 AQI Value':'NILAI_AQI_CO2', 'NO2 AQI Category':'KATEGORI_AQI_CO2', 'PM2.5 AQI Value':'NILAI_AQI_PM2.5',
       'PM2.5 AQI Category':'KATEGORI_AQI_PM2.5'}, inplace=True)
data
```

```
data.isnull().sum()
```

```
df = data[['NILAI_AQI','KATEGORI_AQI']]
df
```

```
df.isnull().sum()
```

```
# normalisasi min-max kolom 
# melakukan data transformasi yaitu normalisasi min-max 
# terhadap kolom bertipe numeric ke dalam rentang nilai 0 - 1
from sklearn.preprocessing import MinMaxScaler

data_skaling = df[['NILAI_AQI']]

# scale features
scaler = MinMaxScaler()
model = scaler.fit(data_skaling)
data_skaling = model.transform(data_skaling)

x = df.copy()
x['NILAI_AQI'] = data_skaling[:,0]
x
```

```
# Melakukan one hot encoding pada kolom bertipe kategorikal
x = pd.get_dummies(x, columns = [ 'KATEGORI_AQI'])
x.reset_index().drop('index', axis='columns')
x
```

```
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1, 11), wcss)
plt.title('Metode Elbow')
plt.xlabel('Jumlah cluster')
plt.show()
```

```
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=5)
kmeans.fit(x)
df['kluster'] = kmeans.labels_
print(kmeans.labels_)
print(silhouette_score(x, kmeans.labels_))
```

```
df_kategori_aqi = pd.factorize(df['KATEGORI_AQI'].unique())

kode_kategori_aqi = {}

for i in range(len(df_kategori_aqi[0])):
    nama = df_kategori_aqi[1][i]
    kode = df_kategori_aqi[0][i]
    kode_kategori_aqi[nama] = kode

for index, row in data.iterrows():
    data.loc[index, 'KODE KATEGORI_AQI'] = kode_kategori_aqi[row['KATEGORI_AQI']]

kode_kategori_aqi
```

```
warna_cluster = df['kluster'].copy().reset_index().drop('index', axis='columns')

for index, row in warna_cluster.iterrows():
    kluster = row['kluster']
    if(kluster == 0):
        warna = 'black'
    elif(kluster == 1):
        warna = 'red'
    elif(kluster == 2):
        warna = 'green'
    elif(kluster == 3):
        warna = 'grey'
    elif(kluster == 4):
        warna = 'pink'
    elif(kluster == 5):
        warna = 'brown'
    warna_cluster.loc[index, 'data_warna'] = warna;
```

```
import numpy as np

fig = plt.Figure(figsize=(30,10))
plt.subplot(1,1,1)

x = np.array(df['NILAI_AQI'])
y = np.array(df['KATEGORI_AQI'])
plt.scatter(x,y, marker=".", c=warna_cluster['data_warna'])
plt.xlabel('Nilai Indeks Kualitas Udara ')
plt.ylabel('Kategori Indeks Kualitas Udara')
plt.show()
```
